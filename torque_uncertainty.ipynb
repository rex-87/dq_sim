```python
import sympy as sym
from IPython.display import Math, Latex
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize'] = 20, 8
plt.rcParams['legend.fontsize'] = 'xx-large'
plt.rcParams['axes.titlesize'] = 20

i_a = sym.Symbol('i_a')
i_b = sym.Symbol('i_b')
i_c = sym.Symbol('i_c')
theta = sym.Symbol('theta')

def get_matrix_latex(M):
    to_print = r"""
    \begin{{bmatrix}}
    {}
    \end{{bmatrix}}
    """.format(sym.printing.latex(M, mat_delim = ''))
    return to_print

def format_latex(in_expr):
    return sym.printing.latex(in_expr, mat_delim = '', long_frac_ratio  = .5, fold_short_frac = False)

#i_abc = sym.Matrix([[i_a],[i_b],[i_c]])
```

# I. Instantaneous Currents

## Graph

```python
p_ia = sym.cos(theta)
p_ib = sym.cos(theta-2*sym.pi/3)
p_ic = sym.cos(theta-4*sym.pi/3)
p1 = sym.plot(p_ia, line_color = 'blue', show = False, label = '$i_a$')
p2 = sym.plot(p_ib, line_color = 'orange', show = False, label = '$i_b$')
p3 = sym.plot(p_ic, line_color = 'grey', show = False, label = '$i_c$')
p1.append(p2[0])
p1.append(p3[0])
p1.legend = True
p1.xlabel = r'$\Omega_s.t$'
p1.ylabel = r'i($\Omega_s.t$)'
p1.show()
```

 # II. Clarke $\alpha\beta\gamma$-Currents

## Clarke Matrix $K_C$

```python
K_C = sym.Rational(2,3)*sym.Matrix([[1, -sym.Rational(1,2), -sym.Rational(1,2)],[0, sym.sqrt(3)/2, -sym.sqrt(3)/2], [sym.Rational(1,2), sym.Rational(1,2), sym.Rational(1,2)]]);

to_print = r"""
\begin{{equation*}}
K_C = {}
\end{{equation*}}
""".format(get_matrix_latex(K_C))
Latex(to_print)
```

$
\begin{equation*}
    \begin{bmatrix}
        i_{\alpha} \\
        i_{\beta} \\
        i_{\gamma}
    \end{bmatrix} =
    K_C*
    \begin{bmatrix}
        i_a \\
        i_b \\
        i_c
    \end{bmatrix}
\end{equation*}
$

## Graph

```python
p_iabg = sym.simplify(K_C*sym.Matrix([[p_ia],[p_ib],[p_ic]]))

p1 = sym.plot(p_iabg[0], line_color = 'blue', show = False, label = r"$i_{\alpha}$")
p2 = sym.plot(p_iabg[1], line_color = 'orange', show = False, label = r"$i_{\beta}$")
p3 = sym.plot(p_iabg[2], line_color = 'grey', show = False, label = r"$i_{\gamma}$")
p1.append(p2[0])
p1.append(p3[0])
p1.legend = True
p1.xlabel = r'$\Omega_s.t$'
p1.ylabel = r'i($\Omega_s.t$)'
p1.show()
```

# III. $dq\gamma$-Currents

## Park Matrix $K_P$

```python
K_P = sym.Matrix([[sym.cos(theta), sym.sin(theta), 0],[-sym.sin(theta), sym.cos(theta), 0], [0, 0, 1]]);

to_print = r"""
\begin{{equation*}}
K_P = {}
\end{{equation*}}
""".format(get_matrix_latex(K_P))
Latex(to_print)
```

$
\begin{equation*}
    \begin{bmatrix}
        i_d \\
        i_q \\
        i_{\gamma} 
    \end{bmatrix} =
    K_P
    \begin{bmatrix}
        i_{\alpha} \\
        i_{\beta} \\
        i_{\gamma} 
    \end{bmatrix}
    \hspace{1cm}
    \implies
    \begin{bmatrix}
        i_d \\
        i_q \\
        i_{\gamma} 
    \end{bmatrix} =
    K_P*K_C*
    \begin{bmatrix}
        i_a \\
        i_b \\
        i_c
    \end{bmatrix}
\end{equation*}
$

## Graph

```python
p_idqg = sym.simplify(K_P*p_iabg)

p1 = sym.plot(p_idqg[0], line_color = 'blue', show = False, label = r"$i_d$")
p2 = sym.plot(p_idqg[1], line_color = 'orange', show = False, label = r"$i_q$")
p3 = sym.plot(p_idqg[2], line_color = 'grey', show = False, label = r"$i_{\gamma}$")
p1.append(p2[0])
p1.append(p3[0])
p1.legend = True
p1.xlabel = r'$\Omega_s.t$'
p1.ylabel = r'i($\Omega_s.t$)'
p1.show()
```

## Calculation

### Park-Clarke Matrix and Inverse

```python
K_PC = K_P*K_C
K_PC_inv = K_PC.inv()

i_d = sym.Symbol('i_d')
i_q = sym.Symbol('i_q')
i_g = sym.Symbol('i_{\gamma}')
sym_i_dqg = sym.Matrix([[i_d],[i_q],[i_g]])

to_print = r"""
\begin{{equation*}}
K_{{PC}} = 
{}
\end{{equation*}}
\begin{{equation*}}
K^{{-1}}_{{PC}} = K^{{T}}_{{PC}} = 
{}
\end{{equation*}}
""".format(
    get_matrix_latex(sym.simplify(K_PC)),
    get_matrix_latex(sym.simplify(K_PC_inv)),
)
Latex(to_print)
```

### $dq\gamma$ vs. instantaneous

```python
G_a = sym.Symbol('G_a')
O_a = sym.Symbol('O_a')
G_b = sym.Symbol('G_b')
O_b = sym.Symbol('O_b')
G_c = sym.Symbol('G_c')
O_c = sym.Symbol('O_c')

i_abc = sym.Matrix([
    [G_a*i_a + O_a],
    [G_b*i_b + O_b],
    [G_c*i_c + O_c]],
)
i_dqg = K_PC*i_abc

sym_i_abc = K_PC_inv*sym_i_dqg

to_print = r"""
\begin{{equation*}}
    \begin{{bmatrix}}
        i_d \\
        i_q \\
        i_{{\gamma}} 
    \end{{bmatrix}} = 
    {}
\end{{equation*}}
\begin{{equation*}}
    \begin{{bmatrix}}
        i_a \\
        i_b \\
        i_c 
    \end{{bmatrix}} = 
    {}
\end{{equation*}}
""".format(
    get_matrix_latex(sym.simplify(i_dqg)),
    get_matrix_latex(sym.simplify(sym_i_abc)),
)
Latex(to_print)
```

# IV. Electro-magnetic Torque

$$T_e = \frac{3}{2} p_b \left( \psi_m i_q + (L_d - L_q) i_d i_q \right)$$

```python
p_b = sym.Symbol('p_b')
psi_m = sym.Symbol('\psi_m')
L_d = sym.Symbol('L_d')
L_q = sym.Symbol('L_q')

T_e = sym.Rational(3,2)*p_b*(psi_m*i_dqg[1] + (L_d-L_q)*i_dqg[0]*i_dqg[1])

to_print = r"""
\begin{{equation*}}
    T_e = {}
\end{{equation*}}
""".format(
    format_latex(T_e),
)
Latex(to_print)
```

# IV. Partial derivatives

## Calculation

```python
dTe_dia = (sym.diff(T_e, i_a))
dTe_dib = (sym.diff(T_e, i_b))
dTe_dic = (sym.diff(T_e, i_c))
dTe_dGa = (sym.diff(T_e, G_a))
dTe_dGb = (sym.diff(T_e, G_b))
dTe_dGc = (sym.diff(T_e, G_c))
dTe_dOa = (sym.diff(T_e, O_a))
dTe_dOb = (sym.diff(T_e, O_b))
dTe_dOc = (sym.diff(T_e, O_c))
dTe_dth = (sym.diff(T_e, theta))
dTe_dpsim = (sym.diff(T_e, psi_m))
dTe_dLd = (sym.diff(T_e, L_d))
dTe_dLq = (sym.diff(T_e, L_q))
dTe_dpb = (sym.diff(T_e, p_b))
```

# V. Uncertainty Propagation

The measurement uncertainties for the **current** sensors and **position angle** are (should be) known.  
- Uncertainties in **current** measurements (random error) :  
$u(i_a) \hspace{.2cm} u(i_b) \hspace{.2cm} u(i_c)$
- Uncertainties in **current sensor gain** :  
$u(G_a) \hspace{.2cm} u(G_b) \hspace{.2cm} u(G_c)$
- Uncertainties in **current sensor offset** :  
$u(O_a) \hspace{.2cm} u(O_b) \hspace{.2cm} u(O_c)$
- Uncertainty in **position angle** measurement :  
$u(\theta)$
- Uncertainties in **motor parameters** :  
$u(\psi_m) \hspace{.2cm} u(L_d) \hspace{.2cm} u(L_q)$

## Combined uncertainty $u_c$ of $dq\gamma$-Currents
The definition of combined uncertainty is taken from [GUM](https://www.bipm.org/utils/common/documents/jcgm/JCGM_100_2008_E.pdf) 5.1.2

$$
\begin{equation*}
    u^2(T_e) =   
    \left( \frac{\partial T_e}{\partial i_a}.u(i_a) \right)^2+
    \left( \frac{\partial T_e}{\partial i_b}.u(i_b) \right)^2+
    \left( \frac{\partial T_e}{\partial i_c}.u(i_c) \right)^2\\+    
    \left( \frac{\partial T_e}{\partial G_a}.u(G_a) \right)^2+
    \left( \frac{\partial T_e}{\partial G_b}.u(G_b) \right)^2+
    \left( \frac{\partial T_e}{\partial G_c}.u(G_c) \right)^2\\+
    \left( \frac{\partial T_e}{\partial O_a}.u(O_a) \right)^2+
    \left( \frac{\partial T_e}{\partial O_b}.u(O_b) \right)^2+
    \left( \frac{\partial T_e}{\partial O_c}.u(O_c) \right)^2\\+
    \left( \frac{\partial T_e}{\partial \theta}.u(\theta) \right)^2\\+
    \left( \frac{\partial T_e}{\partial \psi_m}.u(\psi_m) \right)^2+
    \left( \frac{\partial T_e}{\partial L_d}.u(L_d) \right)^2+
    \left( \frac{\partial T_e}{\partial L_q}.u(L_q) \right)^2
\end{equation*}
$$

Note: $p_b$ is constant, so
$
\begin{equation*}
    u(p_b) = 0
\end{equation*}
$

## Calculation

```python
u_ia = sym.Symbol(r'u(i_{a})')
u_ib = sym.Symbol(r'u(i_{b})')
u_ic = sym.Symbol(r'u(i_{c})')
u_Ga = sym.Symbol(r'u(G_{a})')
u_Gb = sym.Symbol(r'u(G_{b})')
u_Gc = sym.Symbol(r'u(G_{c})')
u_Oa = sym.Symbol(r'u(O_{a})')
u_Ob = sym.Symbol(r'u(O_{b})')
u_Oc = sym.Symbol(r'u(O_{c})')
u_th = sym.Symbol(r'u(\theta)')
u_psim = sym.Symbol(r'u(\psi_{m})')
u_Ld = sym.Symbol(r'u(L_{d})')
u_Lq = sym.Symbol(r'u(L_{q})')

# common function for each term of the sum
def get_sum_term(partial_d, u):
    partial_d_u = partial_d.multiply_elementwise(sym.Matrix(3*[u]))
    partial_d_u_2 = partial_d_u.multiply_elementwise(partial_d_u)
    return partial_d_u_2

# common function for each term of the sum (torque)
def get_sum_term_2(partial_d, u):
    partial_d_u = partial_d*u
    partial_d_u_2 = partial_d_u**2
    return partial_d_u_2

# the terms of the sum
dTe_dia_u_ia_2 = get_sum_term_2(dTe_dia, u_ia)
dTe_dib_u_ib_2 = get_sum_term_2(dTe_dib, u_ib)
dTe_dic_u_ic_2 = get_sum_term_2(dTe_dic, u_ic)
dTe_dGa_u_Ga_2 = get_sum_term_2(dTe_dGa, u_Ga)
dTe_dGb_u_Gb_2 = get_sum_term_2(dTe_dGb, u_Gb)
dTe_dGc_u_Gc_2 = get_sum_term_2(dTe_dGc, u_Gc)
dTe_dOa_u_Oa_2 = get_sum_term_2(dTe_dOa, u_Oa)
dTe_dOb_u_Ob_2 = get_sum_term_2(dTe_dOb, u_Ob)
dTe_dOc_u_Oc_2 = get_sum_term_2(dTe_dOc, u_Oc)
dTe_dth_u_th_2 = get_sum_term_2(dTe_dth, u_th)
dTe_dpsim_u_psim_2 = get_sum_term_2(dTe_dpsim, u_psim)
dTe_dLd_u_Ld_2 = get_sum_term_2(dTe_dLd, u_Ld)
dTe_dLq_u_Lq_2 = get_sum_term_2(dTe_dLq, u_Lq)

# sum the terms!
u_Te_2 = (
    dTe_dia_u_ia_2 + 
    dTe_dib_u_ib_2 + 
    dTe_dic_u_ic_2 +  
    dTe_dGa_u_Ga_2 +
    dTe_dGb_u_Gb_2 +
    dTe_dGc_u_Gc_2 +
    dTe_dOa_u_Oa_2 +
    dTe_dOb_u_Ob_2 +
    dTe_dOc_u_Oc_2 +
    dTe_dth_u_th_2 +
    dTe_dpsim_u_psim_2 +
    dTe_dLd_u_Ld_2 +
    dTe_dLq_u_Lq_2
)
# square root
u_Te = sym.sqrt(u_Te_2)

to_print = r"""
\begin{{equation*}}
    u(T_e) =
    {}
\end{{equation*}}
""".format(
    get_matrix_latex(u_Te),
)
# Latex(to_print)
```
